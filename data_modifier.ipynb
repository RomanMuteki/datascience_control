```python
import pandas as pd
df = pd.read_csv('data/observ.csv')
df.tail()
```

```python
df1 = df[['image_url', 'scientific_name', 'common_name']].copy()
for index, row in df1.iterrows():
    spec = row['scientific_name'].split(' ')
    spec = spec[0] + ' ' + spec[1]
    row['scientific_name'] = spec
```

```python
df1 = df1.groupby('scientific_name').head(1000).reset_index(drop=True)
print(df1)
df1.to_csv('data/observ_cpt.csv', index=False)
```
