```python
import pandas as pd
df = pd.read_csv('data/observ.csv')
df.tail()
```

```python
df1 = df[['image_url', 'scientific_name', 'common_name']].copy()
for index, row in df1.iterrows():
    spec = row['scientific_name'].split(' ')
    if len(spec) > 1:
        spec = spec[0] + '_' + spec[1]
    else:
        spec = row['scientific_name']
        if "," in spec:
            spec = spec.replace(',', '')
    row['scientific_name'] = spec
```

```python
df1 = df1.groupby('scientific_name').head(600).reset_index(drop=True)
print(df1)
df1.to_csv('data/observ_cpt.csv', index=False)
```
