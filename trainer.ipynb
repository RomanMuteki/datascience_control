```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from sympy import evaluate
import numpy as np
from sklearn.metrics import f1_score
import random
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True


torch.manual_seed(42)
np.random.seed(42)
random.seed(42)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu") 
print("Работаем на", device, "\n") 
```

```python
stats_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

temp_dataset = ImageFolder("./data/train", transform=stats_transform)
temp_loader = DataLoader(temp_dataset, batch_size=64, shuffle=False, num_workers=0)

print("Вычисление mean и std для train датасета ...")
cnt = 0
mean = np.zeros(3)
var = np.zeros(3)

for img, _ in temp_loader:
    b = img.size(0)
    img = img.view(b, img.size(1), -1)
    mean += img.mean(2).sum(0).numpy()
    var += img.var(2).sum(0).numpy()
    cnt += b

mean = mean / cnt
var = var / cnt
std = np.sqrt(var)

print("Mean:", mean)
print("Std: ", std)
```

```python
train_trans = transforms.Compose(
    [
        transforms.Resize((256, 256)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.RandomRotation(degrees=(0, 90)),
        #transforms.ColorJitter(brightness=0.2, contrast=0.2),
        transforms.ToTensor(),
        transforms.Normalize(mean=mean.tolist(), std=std.tolist())
    ]
)

val_trans = transforms.Compose(
    [
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize(mean=mean.tolist(), std=std.tolist())
    ]
)
```

```python
train_dataset = ImageFolder("./data/train", transform=train_trans)
val_dataset = ImageFolder("./data/test", transform=val_trans)

train_loader = DataLoader(train_dataset, batch_size=48, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=48, shuffle=False)
```
